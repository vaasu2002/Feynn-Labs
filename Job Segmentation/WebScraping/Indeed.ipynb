```python
roll = []
loc = []
name=[]
date = []
dec = []
while(True):
  try:
    url ='https://in.indeed.com'+soup.find('a',{'aria-label':'Next'}).get('href')
  except AttributeError:
    break
  code =   requests.get(url)
  soup = BeautifulSoup(code.text,'html.parser')

  for i in soup.find_all('h2',class_='jobTitle'):
    roll.append(i.get_text())

  for j in soup.select('span.companyName'):
    name.append(j.get_text())

  for k in soup.select('div.companyLocation'):
    loc.append(k.get_text())

  for l in soup.select('span.date'):
    date.append(l.get_text())

  for m in soup.select('div.job-snippet'):
    dec.append(m.get_text().strip('\n'))
```

```python
import pandas as pd
df = pd.DataFrame()
df['Date'] = date
df['Company Name'] = name
df['Location'] = loc
df['Job Title'] = roll
df['Description'] = dec
df
```

```python
df.to_csv('Indeed.csv')
```
