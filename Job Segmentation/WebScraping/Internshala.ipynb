```python
url="https://internshala.com/internships/data%20science-internship-in-bangalore"
wanted_list=["Computer Vision","Smart Cricket Global Limited","Bangalore"]
```

```python
scraper=AutoScraper()
result=scraper.build(url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_25ui':'Location','rule_b7ni':'Role','rule_1rt9':'Company'})
scraper.keep_rules(['rule_25ui','rule_b7ni','rule_1rt9'])
scraper.save('url-search')
```

```python
import pandas as pd
```

```python
for i in range(6):
  results=scraper.get_result_similar(f'https://internshala.com/internships/data%20science-internship/page-{i+1}',group_by_alias=True)
  dict = {'Company': results['Company'], 'Location': results['Location'], 'Role': results['Role']}
  df = pd.DataFrame(dict) 
  df.to_csv(f'internshalla{i+1}.csv') 
```

```python
l = []
for i in range(6):
  l.append(f"internshalla{i+1}.csv")
```
