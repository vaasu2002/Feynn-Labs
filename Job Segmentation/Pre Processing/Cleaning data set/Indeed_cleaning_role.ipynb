```python
pattern = ['NLP','Computer Vision','ML Engineer','Data Engineer','Machine Learning Engineer','Analytics','Analysis','Consultant','Data Science Engineer','Software Engineer','Data Scientist','Data Science']
```

```python
for x in pattern:
  for i in range(len(l)):
    m = re.findall(x, l[i]) 
    if(len(m)!=0):
      l[i] = m[0]
```
