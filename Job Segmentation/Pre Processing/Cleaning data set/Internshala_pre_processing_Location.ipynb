## DATA SCIENCE

```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/vaasu2002/Feynn-Labs/main/Job%20Segmentation/DATASET/internshalla_final.csv")
df.head()
```

```python
df.drop("Unnamed: 0",1,inplace=True)
```

```python
df.head()
```

```python
df['Location'].value_counts()
```

```python
d={'Work From Home':'Remote', 'Bangalore':'Karnataka','Navi Mumbai':'Maharastra','Mumbai':'Maharastra','Pune':'Maharastra','Gurgaon':'Delhi NCR','Noida':'Delhi NCR','Delhi':'Delhi NCR','Hyderabad':'Telengana'}
```

```python
name=list(df['Location'].value_counts().index)
name
```

```python
for i in name[8:]:
  if(i not in d):
    d[i] = 'Other'
```

```python
d
```

```python
df['Location']=df['Location'].apply(lambda x:d[x])
df.head()
```

```python
df['Location'].value_counts()
```

```python
df.to_csv('internshalla1.csv')
```
